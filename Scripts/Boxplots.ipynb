```python
def plot_box(dataframe):
    t_list=[]
    f_list=[]
    for i in dataframe.columns:
        t_list.extend([i for t in range(len(dataframe))])
        f_list.extend(dataframe[i])
    df = pd.DataFrame({
    "Model":t_list,
    "AUC":f_list
    })
    return df
```
